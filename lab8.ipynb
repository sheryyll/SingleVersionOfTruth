```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data={
    'Product_ID':[101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112],
    'Name' : [
        'Running Shoes', 'Formal Shoes', 'Yoga Mat', 'Protein Shake', 'Dumbblles', 'Office Chair', 'Mechanical keyboard', 'Gaming Mouse', 'Standing Desk', 
        'Resistance Bands', 'Vitmain C Tablets', 'Leather Loafers'
    ],
    'Features': [
        'Sports Footwear Nike Gym',
        'Formal Footware HushPuppies Office',
        'Fitness Gear Equipment Yoga',
        'Fitness Nutrition Health Supplement',
        'Fitness Gear Equipment Weights',
        'Office Furniture Comfort Ergonomic',
        'Electronics Peripherals Gaming RGB',
        'Electronics Peripherals Gaming Wireless',
        'Office Furniture Home Workspace',
        'Fitness Gear Equiment Strength',
        'Fitness Nutrition Health Wellness',
        'Formal Footware Classic Brown'
    ]
}
```

```python
df=pd.DataFrame(data)
```

```python
vectorizer = CountVectorizer()
feature_matrix =  vectorizer.fit_transform(df['Features'])
```

```python
similarity_matrix = cosine_similarity(feature_matrix)
```

```python
target_id = 107
target_idx = df[df['Product_ID'] == target_id].index[0]
```

```python
all_scores = list(enumerate(similarity_matrix[target_idx]))
```

```python
recommended_indices = sorted(all_scores, key=lambda x: x[1] , reverse=True)[1:4]
```

```python
print(f"---Customer Last Interaction: {df.iloc[target_idx]['Name']}---\n")
print(f"Features Analyzed: {df.iloc[target_idx]['Features']}\n")
print("Top 3 Personalized Recommendations:")

for i, score in recommended_indices:
    name = df.iloc[i]['Name']
    feat = df.iloc[i]['Features']
    match = round(score * 100, 1)
    print(f". {name} | Match: {match}% | (Tags: {feat})")
```
